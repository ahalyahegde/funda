```
%%markdown
list vs collection dequeue 

1. dequeue uses the doubly-linkedlist in the background, hence no data movement required when the large data need to be inserted.
2. list requires moving the data, when there is a large data to be inserted, as list stored the data in contegeous memory location
3. popleft and appendleft is Big-O(1) in deque, where as list append() and pop() are Big-O(n)
```

```
from collections import deque
stack = deque([1,2,3])

stack.popleft()
print(stack)

stack.appendleft(4)
print(stack)

print(stack.peak())



```

```
braces = '{'
open_brackets = ['(', '{', '[']
closed_brackets = [')', '}', ']']
stack = []

for i in braces:
    if i in open_brackets:
        stack.append(i)
    elif i in closed_brackets:
        pos = closed_brackets.index(i)
        if len(stack) > 0 and stack[len(stack) -1] == open_brackets[pos]:
            stack.pop()
            print(f"Balanced {open_brackets[pos]} {i}", )
        else:
            print(f"Unbalanced {i}")
            
if len(stack) == 1:
    print(f"unbalanced {braces}")
        
    
        
        
```
