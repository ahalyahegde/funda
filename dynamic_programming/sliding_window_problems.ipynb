```
text = "pwwkew"
repeating_substr = list()
for i in text:
    if i in repeating_substr:
        repeating_substr.remove(i)
        repeating_substr.append(i)
    else:
        repeating_substr.append(i)
    
print(repeating_substr, len(repeating_substr))
```

```
# Input: N = 4, K = 2 | Arr = [100, 200, 300, 400] | Output: 700 | Explanation: Arr3  + Arr4 =700, which is maximum.
n = 4
k = 2
arr = [100, 200, 300, 400]
max_sum = 0
sums = 0
i = 0
j = 0

while j < n:
    sums += arr[j]
    
    if j-i+1 == k:
        max_sum = max(max_sum, sums)
        sums -= arr[i]
        i += 1
    j += 1
    
print(max_sum)
```
